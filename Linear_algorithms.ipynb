# Linear regression

## Fit the data:

* **Residual:** The distance from a line to a data point.
* The purpose is to find the minimum square. Then we rotate the line and sum of the squared residuals â†’ find the rotation that has the least sum of squares.
* y = w1 + w2*x (This is the equation for the line), Least-squares estimated two parameters:
  * y: Intercept
  * x: Slope

image.png

### The square loss

- Average of (Residual)^2 (1/n (Residual_1 + residual_2 + ...)

Caculation:
* **Step 1**: Calculate the average of the intercept â†’ find the mean

* **Step 2**: Some of the squared residuals (the distance from the data point to the mean) â†’ **SS(mean) sum of squares around the mean**

  - SS(mean) = (data - mean)^2
  - Variation around the mean = SS(mean) / n

* **Step 3:** Calculate **SS(fit)**

- SS(fit) = (data - line)^2 (SS(fit) ~ Square loss)
- Var(fit) = ss(fit) / n (var(fit) ~ Training loss)

## Optimization

Gradient: is the direction that increase the training loss the most - Build gradient descent

image.png

Gradient Descent of TrainLoss(w) = 1/n * sum of 2*(predict - label) * feature vector
### Gradient descent
We try to find the minimum of the diravitive of the TrainLoss function
- Step 1: Caculate the diravitive of the TrainLoss function with initial gradient w[0,0]
- Step 2: Update gradient w = w - 0.1(trainLoss)'
- Step 3: Return to step 1 but use the new gradient value w



## Implement code

```python
import numpy as np

# Declare the training dataset and two initial function
trainExample = [
    (1,1), (2,3), (4,3)
]

def phi(x):
  return np.array([1, x])

def initialWeightVector():
  return np.zeros(2)


# Trainloss function, gradientTrailoss and gradientDescent
def trainloss(w): # Function
  return (1.0 / len(trainExample))  *  sum((w.dot(phi(x)) - y)**2 for x,y in trainExample)

def gradientTrainloss(w): # Gradient or Derivative of function
  return (1.0 / len(trainExample))  *  sum(2*(w.dot(phi(x)) - y)*phi(x) for x,y in trainExample)

def gradientDescent(F, gradientF, initialWeightVector): # Gradient descent is the algorithms that find the local minimum value of function
  # F: trainloss, gradientF: gradientTrainloss
  w = initialWeightVector()
  eta = 0.1
  # Main loop
  for i in range(501):
    value = F(w) # Value of trainLoss function
    gradient = gradientF(w)
    w = w - eta * gradient
    print(f'epoch {i}: Weight Vector = {w}, TrainLoss{w} = {value}, gradient of trainLoss = {gradient}')

gradientDescent(trainloss, gradientTrainloss, initialWeightVector)
```
